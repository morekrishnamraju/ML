```python
x=df['YearsExperience'].values
y=df['Salary'].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
import numpy as np
X=np.array(x)
Y=np.array(y)
X
```

```python
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
X
```

```python
Y
```

```python
X_min=X.min()
X_max=X.max()
X_norm=(X-X_min)/(X_max-X_min)
X_norm
```

```python
m=np.random.randn()
c=np.random.randn()
print('Randomly initialized slope:',m)
print('Randomly initialized intercept:',c)
```

```python
def predict(x,m,c):
  y_pred=m*X+c
  return y_pred
```

```python
y_pred=predict(X_norm,m,c)
y_pred
```

```python
print('First 5 predicted values:')
print(y_pred[:5])
```

```python
def compute_cost(y,y_pred):
  n=len(y)
  cost=(1/ (2*n))* np.sum((y_pred-y)**2)
  return cost
```

```python
cost=compute_cost(y,y_pred)
print('Initail Cost(MSE):',cost)
```

```python
m=0.0
c=0.0
y_pred=predict(X_norm,m,c)
cost=compute_cost(Y,y_pred)
print('Cost for m=',m,'and c=',c,'is:',cost)
```

```python
learning_rate=0.01
epochs=1000
print('learning rate',learning_rate)
print('number of epochs',epochs)
```

```python
n=len(y)
cost_history=[]
```

```python
for epoch in range(epochs):
  y_pred=predict(X_norm,m,c)
  dm=(1 / n)* np.sum((y_pred-y)*X_norm)
  dc=(1 / n)* np.sum((y_pred-y))
  m=m-learning_rate*dm
  c=c-learning_rate*dc
  cost=compute_cost(y,y_pred)
  cost_history.append(cost)

  if epoch%100==0:
    print('Epoch:',epoch,'Cost:',cost)
```

```python
final_slope=m
final_intercept=c
print('Final Model Parameters:')
print('Final slope(m):',final_slope)
print('Final intercept(c):',final_intercept)
#
```

```python
def predict_salary(years_experience,m,c,X_norm):
  salary_pred=m*X_norm+c
  return salary_pred
```

```python
years=5
predicted_salary=predict_salary(years,final_slope,final_intercept,X_norm)
print(f'Predicted salary for {years} years of experience:',predicted_salary[0][0])
```

```python
y_pred_final=predict(X_norm,final_slope,final_intercept)
mse=np.mean((y-y_pred_final)**2)
print('Mean Squared Error(MSE):',mse)
```

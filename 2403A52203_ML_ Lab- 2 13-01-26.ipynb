```python
age_array=ds['Age'].to_numpy()
age_array
```

```python
age_array=ds['Age'].to_numpy().reshape(-1,1)
age_array
```

```python
mean_age=np.mean(age_array)
median_age=np.median(age_array)
std_age=np.std(age_array)
print(mean_age,',',median_age,',',std_age)
```

```python
cholestrol_array=ds['Cholesterol'].to_numpy()
cholestrol_max=np.max(cholestrol_array)
cholestrol_min=np.min(cholestrol_array)
print("Maximum Cholestrol:",cholestrol_max)
print("Minimum Cholestrol:",cholestrol_min)
```

```python
heart_disease_count=ds['Heart Disease'].value_counts()
heart_disease_count
```

```python
import matplotlib.pyplot as plt
plt.hist(age_array,bins=100)
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Age Distribution')
plt.show()
```

```python
value,counts=np.unique(cholestrol_array,return_counts=True)
plt.bar(value,counts)
plt.xlabel('Cholestrol Level')
plt.ylabel('Number of Patients')
plt.title('Bar Diagram of cholestrol Levels')
plt.show()
```

```python
plt.scatter(ds['Age'],ds['Max HR'])
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.title('Scatter Plot of Age vs Maximum Heart Rate')
plt.show()
```
